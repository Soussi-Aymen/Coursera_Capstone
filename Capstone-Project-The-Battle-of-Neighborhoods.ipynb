# Famous places in Tunisia found in the data

```python
tunisia_venues = tunisia_results['response']['venues']
tunisia_dataframe = json_normalize(tunisia_venues)


tunisia_filtered_columns = ['name', 'categories'] + [col for col in tunisia_dataframe.columns if col.startswith('location.')] + ['id']
tunisia_dataframe_filtered = tunisia_dataframe.loc[:, tunisia_filtered_columns]

def get_category_type(row):
    try:
        categories_list = row['categories']
    except:
        categories_list = row['venue.categories']
        
    if len(categories_list) == 0:
        return None
    else:
        
        return categories_list[0]['name']
tunisia_dataframe_filtered['categories'] = tunisia_dataframe_filtered.apply(get_category_type, axis=1) 

tunisia_dataframe_filtered.columns = [column.split('.')[-1] for column in tunisia_dataframe_filtered.columns]

print(tunisia_dataframe['name'].head(20))
tunisia_dataframe_filtered.head()
```

## Building a map based on Tunis city center and containing all coworking spaces foung
__Note: The green cercle represent Tunisia center and the other cercles represents blue coworking spaces__

```python

venues_map = folium.Map(location=[latitude, longitude], zoom_start=13) # generate map centred around Tunis city center


folium.features.CircleMarker(
    [latitude, longitude],
    radius=10,
    color='green',
    popup='Tunisia City Center',
    fill = True,
    fill_color = 'green',
    fill_opacity = 0.6
).add_to(venues_map)


for lat, lng, label in zip(coworking_dataframe_filtered.lat, coworking_dataframe_filtered.lng, coworking_dataframe_filtered.name):
    folium.features.CircleMarker(
        [lat, lng],
        radius=5,
        color='#2e4a62',
        popup=label,
        fill = True,
        fill_color='#2e4a62',
        fill_opacity=0.6
    ).add_to(venues_map)

venues_map
```
